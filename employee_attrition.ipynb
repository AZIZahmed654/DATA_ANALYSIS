## Data Visualization

Visualize the data to uncover patterns and trends.


## Histogram of Employee Age

In this section, we visualize the distribution of employee ages within the dataset. The histogram, along with a Kernel Density Estimate (KDE), provides insights into the age distribution and density of employees in the company. This helps in understanding the demographic profile of the workforce.


```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set visualization style
sns.set(style="whitegrid")

# Histogram of employee age
plt.figure(figsize=(10, 6))
sns.histplot(df['Age'], kde=True, bins=20)
plt.title('Distribution of Employee Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()


```

## Bar Plot of Attrition by Department

This bar plot shows the distribution of employee attrition across different departments. By visualizing the count of employees who have left versus those who have stayed, we can identify which departments have higher attrition rates. This information is crucial for targeted interventions and improving employee retention in specific departments.


```python
# Bar plot of Attrition by Department
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Department', hue='Attrition')
plt.title('Attrition by Department')
plt.xlabel('Department')
plt.ylabel('Count')
plt.show()
```

## Attrition vs. Age

This box plot illustrates the relationship between employee attrition and age. By comparing the age distributions of employees who have left versus those who have stayed, we can identify whether age is a significant factor influencing attrition. This visualization helps in understanding if certain age groups are more prone to leaving the company.


```python
# Attrition vs. Age
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Attrition', y='Age')
plt.title('Attrition vs. Age')
plt.xlabel('Attrition')
plt.ylabel('Age')
plt.show()

```

## Attrition by Job Role

This bar plot displays the distribution of employee attrition across different job roles. By visualizing the count of employees who have left versus those who have stayed in each job role, we can identify which roles have higher attrition rates. This analysis helps in pinpointing specific roles that may require more focused retention strategies.


```python
# Attrition by Job Role
plt.figure(figsize=(12, 8))
sns.countplot(data=df, x='JobRole', hue='Attrition')
plt.title('Attrition by Job Role')
plt.xlabel('Job Role')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

## Attrition vs. Distance from Home

This box plot shows the relationship between employee attrition and the distance from home. By comparing the distances employees travel to work, we can identify if longer commutes are associated with higher attrition rates. This visualization helps in understanding whether distance from home is a significant factor in employee turnover.


```python
# Attrition vs. Distance from Home
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Attrition', y='DistanceFromHome')
plt.title('Attrition vs. Distance from Home')
plt.xlabel('Attrition')
plt.ylabel('Distance from Home')
plt.show()
```

```python
# Calculate attrition rates by education field
education_attrition = df.groupby('EducationField')['Attrition'].value_counts(normalize=True).unstack() * 100

# Rename columns for clarity
education_attrition.columns = ['No Attrition (%)', 'Attrition (%)']

# Reset index for better readability in the plot
education_attrition = education_attrition.reset_index()

education_attrition
```

## Attrition Rates by Education Field

This bar plot visualizes the attrition rates across different education fields. By comparing the percentage of employees who have left versus those who have stayed within each education field, we can identify if certain educational backgrounds are more associated with higher attrition rates. This analysis provides insights into whether education field influences employee retention.


```python
# Melt the dataframe for better visualization
education_attrition_melted = education_attrition.melt(id_vars='EducationField', value_vars=['No Attrition (%)', 'Attrition (%)'],
                                                      var_name='Attrition Status', value_name='Percentage')

# Plot the data
plt.figure(figsize=(12, 8))
sns.barplot(data=education_attrition_melted, x='EducationField', y='Percentage', hue='Attrition Status')
plt.title('Attrition Rates by Education Field')
plt.xlabel('Education Field')
plt.ylabel('Percentage')
plt.xticks(rotation=45)
plt.show()
```

```python
# Calculate attrition rates by department
department_attrition = df.groupby('Department')['Attrition'].value_counts(normalize=True).unstack() * 100

# Rename columns for clarity
department_attrition.columns = ['No Attrition (%)', 'Attrition (%)']

# Reset index for better readability in the plot
department_attrition = department_attrition.reset_index()

department_attrition
```

## Attrition Rates by Department

This bar plot illustrates the attrition rates across different departments. By comparing the percentage of employees who have left versus those who have stayed within each department, we can identify which departments experience higher attrition rates. This visualization helps in understanding if specific departments have higher turnover, which can guide targeted retention efforts.


```python
# Melt the dataframe for better visualization
department_attrition_melted = department_attrition.melt(id_vars='Department', value_vars=['No Attrition (%)', 'Attrition (%)'],
                                                       var_name='Attrition Status', value_name='Percentage')

# Plot the data
plt.figure(figsize=(12, 8))
sns.barplot(data=department_attrition_melted, x='Department', y='Percentage', hue='Attrition Status')
plt.title('Attrition Rates by Department')
plt.xlabel('Department')
plt.ylabel('Percentage')
plt.xticks(rotation=45)
plt.show()
```
