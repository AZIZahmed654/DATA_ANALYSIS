## Data Preprocessing

In this section, we perform essential preprocessing tasks to prepare the data for analysis. 

### Checking for Missing Values

First, we need to check for any missing values in the dataset. Missing values can impact our analysis and might need to be addressed appropriately. The following code calculates the number of missing values for each column:



````python
# Check for missing values
data.isnull().sum()

# Convert 'Order Date' and 'Ship Date' to datetime format
data['Order Date'] = pd.to_datetime(data['Order Date'])
data['Ship Date'] = pd.to_datetime(data['Ship Date'])

# Display the updated dataframe info
data.info()
````

## Distribution of Sales by Region

In this section, we visualize the distribution of sales across different regions to understand how sales are distributed geographically.

### Visualization

We use a horizontal bar plot to depict the count of sales transactions for each region. This visualization helps us quickly identify which regions have the highest and lowest number of sales.

The following code snippet creates the plot:




````python
import matplotlib.pyplot as plt
import seaborn as sns

# Distribution of sales by region
plt.figure(figsize=(10, 6))
sns.countplot(y='Region', data=data, order=data['Region'].value_counts().index)
plt.title('Distribution of Sales by Region')
plt.xlabel('Number of Sales')
plt.ylabel('Region')
plt.show()

````

## Relationship Between Item Type and Total Profit

In this section, we analyze the relationship between item types and the total profit generated from each type. This helps us understand which items are the most profitable and their impact on overall profitability.

### Visualization

We use a horizontal bar plot to display the total profit for each item type. This visualization provides a clear comparison of profitability across different item categories.

The following code snippet generates the plot:



````python
# Relationship between item type and total profit
plt.figure(figsize=(12, 8))
sns.barplot(x='Total Profit', y='Item Type', data=data, estimator=sum, ci=None)
plt.title('Total Profit by Item Type')
plt.xlabel('Total Profit')
plt.ylabel('Item Type')
plt.show()
````

## Trend of Total Revenue Over Time

In this section, we analyze how total revenue has changed over time. Understanding revenue trends can help identify seasonal patterns, growth trends, or any anomalies.

### Visualization

We use a line plot to visualize the trend of total revenue over time. This type of plot is ideal for showing how a metric evolves over a continuous period.

The following code snippet creates the plot:



````python
# Trend of total revenue over time
plt.figure(figsize=(14, 7))
data.groupby('Order Date')['Total Revenue'].sum().plot()
plt.title('Trend of Total Revenue Over Time')
plt.xlabel('Order Date')
plt.ylabel('Total Revenue')
plt.show()
````

## Total Profit by Country

In this section, we analyze the total profit generated by each country. This analysis helps identify which countries contribute the most to overall profitability and provides insights into regional performance.

### Visualization

We use a horizontal bar plot to display the total profit for each country. This visualization allows for an easy comparison of profitability across different countries.

The following code snippet generates the plot:



````python
# Total profit by country
plt.figure(figsize=(14, 10))
profit_by_country = data.groupby('Country')['Total Profit'].sum().sort_values(ascending=False)
sns.barplot(x=profit_by_country, y=profit_by_country.index)
plt.title('Total Profit by Country')
plt.xlabel('Total Profit')
plt.ylabel('Country')
plt.show()
````

## Most Sold Item per Country

In this section, we identify and visualize the item that is sold the most in each country. This analysis helps us understand the most popular items across different regions.

### Analysis

1. **Group by Country and Item Type:** We start by grouping the data by 'Country' and 'Item Type' to sum the total units sold for each item type within each country. This aggregation allows us to assess sales performance at a more granular level.

    ```python
    sales_by_country_item = data.groupby(['Country', 'Item Type'])['Units Sold'].sum().reset_index()
    ```

2. **Find the Most Sold Item per Country:** From the aggregated data, we determine which item type has the highest total units sold per country. This involves identifying the item with the maximum units sold for each country.

    ```python
    most_sold_items = sales_by_country_item.loc[sales_by_country_item.groupby('Country')['Units Sold'].idxmax()]
    ```

### Visualization

We use a bar plot to illustrate the most sold item in each country. The plot displays the number of units sold for the top item in each country.

The following code snippet generates the plot:



````python
# Group by Country and Item Type, then sum the Units Sold
sales_by_country_item = data.groupby(['Country', 'Item Type'])['Units Sold'].sum().reset_index()

# Find the most sold item per country
most_sold_items = sales_by_country_item.loc[sales_by_country_item.groupby('Country')['Units Sold'].idxmax()]

# Plot
plt.figure(figsize=(12, 8))
for country in most_sold_items['Country'].unique():
    subset = most_sold_items[most_sold_items['Country'] == country]
    plt.bar(subset['Country'], subset['Units Sold'], label=subset['Item Type'].values[0])

plt.xlabel('Country')
plt.ylabel('Units Sold')
plt.title('Most Sold Item per Country')
plt.xticks(rotation=90)
plt.legend(title='Item Type')
plt.tight_layout()
plt.show()
````

## Item with Most Revenue per Country

In this section, we analyze and visualize which item type has generated the highest total revenue in each country. This helps us understand which products are the most profitable in different regions.

### Analysis

1. **Group by Country and Item Type:** We begin by grouping the data by 'Country' and 'Item Type' and then summing the total revenue for each item type within each country. This aggregation provides a clear view of how revenue is distributed across different items and countries.

    ```python
    revenue_by_country_item = data.groupby(['Country', 'Item Type'])['Total Revenue'].sum().reset_index()
    ```

2. **Find the Item with the Most Revenue per Country:** From the aggregated data, we determine which item type has generated the highest total revenue in each country. This involves identifying the item with the maximum revenue for each country.

    ```python
    most_revenue_items = revenue_by_country_item.loc[revenue_by_country_item.groupby('Country')['Total Revenue'].idxmax()]
    ```

### Visualization

We use a bar plot to visualize the item with the highest total revenue in each country. The plot displays the revenue generated by the top item in each country.

The following code snippet generates the plot:



````python
# Group by Country and Item Type, then sum the Total Revenue
revenue_by_country_item = data.groupby(['Country', 'Item Type'])['Total Revenue'].sum().reset_index()

# Find the item with the most revenue per country
most_revenue_items = revenue_by_country_item.loc[revenue_by_country_item.groupby('Country')['Total Revenue'].idxmax()]

# Plot
plt.figure(figsize=(12, 8))
for country in most_revenue_items['Country'].unique():
    subset = most_revenue_items[most_revenue_items['Country'] == country]
    plt.bar(subset['Country'], subset['Total Revenue'], label=subset['Item Type'].values[0])

plt.xlabel('Country')
plt.ylabel('Total Revenue')
plt.title('Item with Most Revenue per Country')
plt.xticks(rotation=90)
plt.legend(title='Item Type')
plt.tight_layout()
plt.show()
````
